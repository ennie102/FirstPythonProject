```python
def student_report(name, marks1, marks2, marks3):
    total = marks1 + marks2 + marks3
    average = total / 3

    if average >= 75:
        grade = "A"
    elif average >= 60:
        grade = "B"
    elif average >= 50:
        grade = "C"
    else:
        grade = "F"

    return f"Student: {name}\nTotal: {total}\nAverage: {average:.2f}\nGrade: {grade}"

    print(student_report("Lamor", 80, 74, 91))
```

```python
#Function calling another function
#Function to calculate area of a circle and print for

import math

def area_of_circle(radius):
    return math.pi * radius ** 2

def display_area(name, radius):
    area = area_of_circle(radius)
    print(f"{name}, the area of your circle with radius {radius} is {area:.2f} square units.")

    #Example calls
    display_area("Avuya", 5)
    display_area("Ennie", 10)

    #Concepts

    #One function (display_area) calls another (area_of_circle).
    #Demonstrates modular programming - breaking code into small, reusable parts
    #Uses the math library for the constant pie
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats

#Dataset
marks = [56,78,45,89,23,78,78,98,80,12,23,44,73,14,67,45,13,77,43,60]

#calculations
mean_val=stats.mean(marks)
median_val=stats.median(marks)
mode_val=stats.mode(marks)
range_val=max(marks)-min(marks)
std_val=stats.stdev(marks)
variance_val=stats.variance(marks)

#Store the results
results = {
    "Mean": mean_val,
    "Median": median_val,
    "Mode": mode_val,
    "Range": range_val,
    "Standard Deviation": std_val,
    "Variance": variance_val
}

#Plot the histogram
plt.hist(marks, bins=5, color = 'green',edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Histogram of student marks')
plt.grid(axis='y' , linestyle= '--',alpha=0.7)
plt.show()

results
```

```python
#BOX PLOT
plt.boxplot(marks,vert=False)
plt.ylabel('Marks')
plt.title('Box Plot of student marks')
plt.grid(axis='x' , linestyle= '--',alpha=0.7)
plt.show()
```
